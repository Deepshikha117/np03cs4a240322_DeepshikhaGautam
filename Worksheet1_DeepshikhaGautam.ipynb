Problem - 1: Array Creation:
Complete the following Tasks:
1.	Initialize an empty array with size 2X2.
2.	Initialize an all one array with size 4X2.
3.	Return a new array of given shape and type, filled with fill value.{Hint: np.full}
4.	Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}
5.	Return a new array of ones with same shape and type as a given array.{Hint: np.ones like}
6.	For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}


```python
import numpy as np
#Task 1
#Empty array 2x2
arr1 = np.empty((2, 2))
print("Empty:\n", arr1)

#All ones 4x2
arr2 = np.ones((4, 2))
print("\nOnes:\n", arr2)

#Fill with specific value 7
arr3 = np.full((2, 2), 7)
print("\nFilled with 7:\n", arr3)

#Zeros with same shape as arr2
arr4 = np.zeros_like(arr2)
print("\nZeros like arr2:\n", arr4)

#Ones with same shape as arr1
arr5 = np.ones_like(arr1)
print("\nOnes like arr1:\n", arr5)

#Convert list to array
my_list = [1, 2, 3, 4]
arr6 = np.array(my_list)

#Display result
print("\nList to Array:", arr6)

```

Problem - 4: Matrix Operations: • For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]); Prove following with Numpy:
1.	Prove A.A−1 = I.
2.	Prove AB ̸= BA.
3.	Prove (AB)T = BTAT.
•	Solve the following system of Linear equation using Inverse Methods.
2x − 3y + z = −1 x − y + 2z = −3 3x + y − z = 9
{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
•	Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function of Numpy}


```python
#Task4
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Linear Equations
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

#Prove Inverse
inv_A = np.linalg.inv(A)
print("A * Inv(A) (Should be Identity):\n", np.dot(A, inv_A).round())

#Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nIs AB equal to BA?", np.array_equal(AB, BA))

#Prove Transpose rule
print("\n(AB) Transpose matches B.T * A.T?", np.array_equal(AB.T, np.dot(B.T, A.T)))

coeffs = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
answers = np.array([-1, -3, 9])

#Solving
xyz = np.linalg.solve(coeffs, answers)

#Displaying result
print("\nSolution for x, y, z:", xyz)
```
